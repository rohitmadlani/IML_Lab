Write a Pandas program to create a plot of Open, High, Low, Close, Adjusted Closing prices and Volume of given company between two specific dates.

```python
import datetime as dt
import matplotlib.pyplot as plt
from matplotlib import style
import pandas as pd
# import yfinance as yf
import plotly.graph_objects as go

start = dt.datetime(2000, 1, 3)
end = dt.datetime(2000, 12, 29)

#df = yf.download(tickers="TSLA", start=start, end = end)
df = pd.read_csv('/kaggle/input/nifty50-stock-market-data/TITAN.csv')
df.reset_index(inplace=True)

df['Date'] = pd.to_datetime(df['Date'])

fig = go.Figure(data=[go.Candlestick(x=df['Date'],
                open=df['Open'],
                high=df['High'],
                low=df['Low'],
                close=df['Close'])])

fig.show()
```
