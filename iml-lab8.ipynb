Write a Pandas program to create a line plot of the opening, closing stock prices of given company between two specific dates.

```python
import datetime as dt
import matplotlib.pyplot as plt
from matplotlib import style
import pandas as pd
# import yfinance as yf

start = dt.datetime(2000, 1, 3)
end = dt.datetime(2000, 12, 29)

#df = yf.download(tickers="TSLA", start=start, end = end)
df = pd.read_csv('/kaggle/input/nifty50-stock-market-data/TITAN.csv')
df.reset_index(inplace=True)

df['Date'] = pd.to_datetime(df['Date'])

new_df = (df['Date']>= start) & (df['Date']<= end)

df2 = df.loc[new_df]

df2.plot(x='Date', y=['Open', 'Close']);
plt.xlabel("Date",fontsize=12, color='black')
plt.ylabel("Price", fontsize=12, color='black')
plt.show()
```
